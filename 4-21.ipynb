```python
weight = 0.5
goal_pred = 0.8
input = 0.5

for iteration in range(20):
    pred = input * weight
    error = (pred - goal_pred) ** 2
    delta = pred - goal_pred

    # 微分係数
    weight_delta = input * delta
    weight = weight - weight_delta

    print(f"[{iteration}] Error: {error}, Prediction: {pred}, Weight: {weight}")
```
