```python
weight = 0.5
goal_pred = 0.8
input = 2.0
alpha = 0.1  # 学習率

for iteration in range(20):
    pred = input * weight
    error = (pred - goal_pred) ** 2
    delta = pred - goal_pred

    # 微分係数
    weight_delta = input * delta
    weight = weight - alpha * weight_delta

    print(f"[{iteration}] Error: {error}, Prediction: {pred}, Weight: {weight}")
```

```python
!pip install matplotlib
import matplotlib.pyplot as plt

# 既存の変数（alpha, input, goal_pred）を使って学習過程を再現
initial_weight = 0.5
w = initial_weight

iterations = []
pred_history = []
error_history = []
weight_history = []

for i in range(20):
    pred_i = input * w
    error_i = (pred_i - goal_pred) ** 2
    delta_i = pred_i - goal_pred
    weight_delta_i = input * delta_i
    w = w - alpha * weight_delta_i

    iterations.append(i)
    pred_history.append(pred_i)
    error_history.append(error_i)
    weight_history.append(w)

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

axes[0].plot(iterations, pred_history, marker="o")
axes[0].axhline(goal_pred, color="r", linestyle="--", label="goal_pred")
axes[0].set_title("Prediction")
axes[0].set_xlabel("Iteration")
axes[0].set_ylabel("pred")
axes[0].legend()

axes[1].plot(iterations, error_history, marker="o")
axes[1].set_title("Error")
axes[1].set_xlabel("Iteration")
axes[1].set_ylabel("error")
axes[1].set_yscale("log")

axes[2].plot(iterations, weight_history, marker="o")
axes[2].set_title("Weight")
axes[2].set_xlabel("Iteration")
axes[2].set_ylabel("weight")

plt.tight_layout()
plt.show()
```
