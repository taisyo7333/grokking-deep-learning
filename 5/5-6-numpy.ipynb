```python
import numpy as np

weights = np.array([
    # toes,wlrec,nfans
    [0.1, 0.1, -0.3], # hurt?
    [0.1, 0.2, 0.0],  # win?
    [0.0, 1.3, 0.1]   # sad?
])

# Input data
toes = [8.5, 9.5, 9.9, 9.0]
wlrec = [0.65, 0.8, 0.8, 0.9]
nfans = [1.2, 1.3, 0.5, 1.0]

# goal predictions
hurt = [0.1, 0.0, 0.0, 0.1]
win = [1, 1, 0, 1]
sad = [0.1, 0.0, 0.1, 0.2]

alpha = 0.01

input = np.array([toes[0], wlrec[0], nfans[0]])
goal_pred = np.array([hurt[0], win[0], sad[0]])

# Inference
pred = np.dot(input, weights.T)
print("input=" + str(input))
print("weights=" + str(weights))
print("pred=" + str(pred))

error = (pred - goal_pred) ** 2
print("error=" + str(error))

delta = pred - goal_pred
print("delta=" + str(delta))

# 微分係数：3x3 matrix
# 各要素は、各weightsの要素に対する誤差の変化率を表す
derivative = np.outer(delta, input)
print("derivative=" + str(derivative))

# update weights
weights = weights - alpha * derivative
print("weights=" + str(weights))
```
