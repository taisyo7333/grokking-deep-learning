5.5 複数の出力を持つ勾配降下法による学習

入力：１
出力：３

```python
import numpy as np
weights = np.array([ 0.3, 0.2, 0.9 ])
wlrec = np.array([ 0.65, 1.0, 1.0, 0.9 ])
hurt = np.array([0.1, 0.0, 0.0, 0.1])
win = np.array([ 1, 1, 0, 1])
sad = np.array([0.1, 0.0, 0.1, 0.2 ])
alpha = 0.1

# input
input = wlrec[0]
# predicted output
truth = np.array([hurt[0], win[0], sad[0]])

def neural_network(input, weights):
    pred = np.dot(input, weights)
    return pred

pred = neural_network(input, weights)

error = (pred - truth) ** 2
delta = pred - truth

print("Predicted:", pred)
print("error:", error)
print("delta:", delta)
print("weights:", weights)

# scalar * vector
weight_delta = np.dot(input, delta)
weights = weights - alpha * weight_delta

print("Weight delta:", weight_delta)
print("Updated weights:", weights)
```

```python
# scalar と　vectorの外積は、内積と同じ結果になる（ただし、外積は行列の形になる）
np.outer(input, delta)
```
