```python
import numpy as np
a = np.array([1,2,3])
b = np.array([4,5,6])

c = a * b # 各要素ごとに掛け算する
c
```

```python
c = a.dot(b) # 内積を計算する
c
```

```python
a = np.array([[1, 2, 3], [4, 5, 6]])
b = np.array([[1, 2], [3, 4], [5, 6]])
print(f"a({a.shape})=\n{a}")
print(f"b({b.shape})=\n{b}")
c = a.dot(b)
c
```
