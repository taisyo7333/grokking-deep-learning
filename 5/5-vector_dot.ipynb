```python
import numpy as np
a = np.array([1,2,3])
b = np.array([4,5,6])

c = a * b # 各要素ごとに掛け算する
c
```

```python
a = np.array([[1, 2, 3], [4, 5, 6]])
b = np.array([[1, 2], [3, 4], [5, 6]])
print(f"a({a.shape})=\n{a}")
print(f"b({b.shape})=\n{b}")
c = a.dot(b)
c
```

```python
import numpy as np
a = np.array([1, 2, 3, 4])
b = np.array([5,6,7])

x = np.outer(a,b)
y = np.outer(a.T,b)

print(f"x({x.shape})=\n{x}")
print(f"y({y.shape})=\n{y}")
```

```python
import numpy as np

# layer: 1x4 matrix
layer = np.array([[0.0,0.51828245,0.0,0.0]])
print(f"layer({layer.shape})=\n{layer}")

# delta: 1x4 matrix
delta = np.array([[ 0.0, -0.45983371,0.0,0.0]])
print(f"delta({delta.shape})=\n{delta}")

# layer.T: 4x1 matrix
print(f"layer.T({layer.T.shape})=\n{layer.T}")

# 4x1 matrix x 1x4 matrix = 4x4 matrix
derivative = layer.T.dot(delta)
print(f"derivative({derivative.shape})=\n{derivative}")



```
