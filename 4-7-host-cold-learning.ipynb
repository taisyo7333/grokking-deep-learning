```python
weight = 0.5
input = 0.5
goal_prediction = 0.8
step_amount = 0.1 # イテレーションごとの重みの変化量

for iteration in range(1101): # 学習を繰り返しながら誤差を小さくしている
    prediction = input * weight # 予測値の計算
    error = (prediction - goal_prediction) ** 2 # 誤差の計算

    print("Error:" + str(error) + " Prediction:" + str(prediction))

    # 大きくする
    up_prediction = input * (weight + step_amount)
    up_error = (goal_prediction - up_prediction) ** 2

    # 小さくする
    down_prediction = input * (weight - step_amount)
    down_error = (goal_prediction - down_prediction) ** 2

    if(down_error < up_error): # weightを小さくした方が良い場合は小さくする
        weight = weight - step_amount

    if(down_error > up_error): # weightを大きくした方が良い場合は大きくする
        weight = weight + step_amount
```
