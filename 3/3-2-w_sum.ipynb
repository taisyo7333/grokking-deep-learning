```python
def w_sum(a, b):
    assert(len(a) == len(b))
    output = 0
    for i in range(len(a)):
        output += a[i] * b[i]
    return output
```

```python
a = [0, 1, 0, 1]
b = [1, 0, 1, 0]
c = [0, 1, 1, 0]
d = [.5, 0, .5, 0]
e = [0, 1, -1, 0]
```

```python
print("w_sum(a, b) = ",w_sum(a, b))
print("w_sum(b, c) = ",w_sum(b, c))
print("w_sum(b, d) = ",w_sum(b, d))
print("w_sum(c, c) = ",w_sum(c, c))
print("w_sum(d, d) = ",w_sum(d, d))
print("w_sum(c, e) = ",w_sum(c, e))
```
