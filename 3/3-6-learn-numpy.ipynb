```python
import numpy as np
```

```python
a = np.array([0,1,2,3]) # Vector
b = np.array([4,5,6,7]) # Vector
c = np.array([
    [0,1,2,3],
    [4,5,6,7]
    ]) # Matrix
d = np.zeros((2,4)) # 2x4 Matrix of zeros
e = np.random.rand(2,5) # 2x5 Matrix of random numbers(from 0 to 1)
```

```python
print(a) #
print(b) #
print(c) #
print(d) #
print(e) #
```

```python
print(a * 0.1)
print(a * b) # Element-wise multiplication
print(a * c) # Element-wise multiplication
# print(a * e)
```

```python
a = np.zeros((1,4))
b = np.zeros((4,3))
c = a.dot(b) # Dot product
print(c)
print(c.shape) # (1,3)
```

```python
a = np.zeros((2,4))
b = np.zeros((4,3))
c = a.dot(b) # Dot product
print(c.shape) # (2,3)
```

```python
e = np.zeros((2,1))
f = np.zeros((1,3))
g = e.dot(f) # Dot product
print(g.shape) # (2,3)
```

```python
h = np.zeros((5,4)).T # Transpose of a 5x4 matrix, resulting in a 4x5 matrix
print(h.shape) # (4,5)

i = np.zeros((5,6))
j = h.dot(i) # Dot product of a 4x5 and 5x6 matrix, resulting in a 4x6 matrix
print(j.shape) # (4,6)
```

```python
h = np.zeros((5,4))
i = np.zeros((5,6))
j = h.T.dot(i) # Dot product of a 4x5 and 5x6 matrix, resulting in a 4x6 matrix
print(j.shape) # (4,6)
```
