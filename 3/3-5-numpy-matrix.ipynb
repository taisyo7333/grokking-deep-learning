```python
import numpy as np
```

```python
# toes wlrec nfans
ih_wgt = np.array([
    [0.1, 0.2, -0.1],  # hid[0] weights
    [-0.1, 0.1, 0.9],  # hid[1] weights
    [0.1, 0.4, 0.1]    # hid[2] weights
]).T
```

# np.array().T
Tは行列の転置を意味します。元の行列は3行3列ですが、転置することで3列3行になります。
これにより、各隠れ層のニューロンに対応する重みが列として配置されます。

```python
# hid[0] hid[1] hid[2]
hp_wgt = np.array([
    [0.3, 1.1, -0.3], # けが？
    [0.1, 0.2, 0.0],  # 勝った？
    [0.0, 1.3, 0.1]   # 悲しい？
]).T
```

```python
weights = [ih_wgt, hp_wgt]
```

```python
def neural_network(input, weights):
    hid = input.dot(weights[0])
    pred = hid.dot(weights[1])
    return pred
```

```python
toes = np.array([8.5, 9.5, 9.9, 9.0])
wlrec = np.array([0.65, 0.8, 0.8, 0.9])
nfans = np.array([1.2, 1.3, 0.5, 1.0])
```

```python
input = np.array([toes[0], wlrec[0], nfans[0]])
```

```python
pred = neural_network(input, weights)
print(pred)
```
